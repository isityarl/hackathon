```python
# комиссии
comm = (
    comissions[['login', 'date_trunc']]
    .rename(columns={'date_trunc': 'date'})
    .assign(event='commission')
)

# торговля
trades_ev = (
    trades[['login', 'date_last_activity']]
    .rename(columns={'date_last_activity': 'date'})
    .assign(event='trade')
)

# ввод/вывод
inouts_ev = (
    inouts
    .merge(acc[['login', 'client_id']], on='client_id', how='left')
    .rename(columns={'date_last_inouts': 'date'})
    .assign(event='inout')
)

# конвертация
conv_ev = (
    exchange
    .merge(acc[['login', 'client_id']], on='client_id', how='left')
    .rename(columns={'date_last_conv': 'date'})
    .assign(event='conversion')
)

all_events = pd.concat([comm, trades_ev, inouts_ev, conv_ev], ignore_index=True)
all_events['date'] = pd.to_datetime(all_events['date'], errors='coerce')

all_events


# for col in ['date_last_activity', 'date_last_inouts', 'date_last_conv', 'date_trunc']:
#     data[col] = pd.to_datetime(data[col], errors='coerce')
```
